# **Data visualization**

```python
number_of_movies_per_year = df[df['type'] == 'Movie'].groupby('release_year')['type'].count()
year_with_most_movies = number_of_movies_per_year.idxmax()
print(year_with_most_movies)
```

```python
number_of_shows_per_year = df[df['type'] == 'TV Show'].groupby('release_year')['type'].count()
year_with_most_shows = number_of_shows_per_year.idxmax()
print(year_with_most_shows)
```

```python
sns.boxplot(
    x = "release_year",
    y = "type",
    showmeans=True,
    data=df
)
plt.xticks(rotation=45)
plt.show()
```

```python
movie_ratings_distribution = df[df['type'] == 'Movie']['rating'].value_counts()
plt.pie(movie_ratings_distribution.values, labels=movie_ratings_distribution.index, autopct="%1.1f%%")
plt.title('Distribution of Movie Ratings')
plt.show()
```

```python
number_of_movies_per_director = df.groupby('director')['show_id'].count()
number_of_movies_per_director = number_of_movies_per_director.sort_values(ascending=False)
top_10_directors = number_of_movies_per_director.index[:10]

number_of_movies_by_top_10_directors = number_of_movies_per_director[top_10_directors]

plt.bar(top_10_directors, number_of_movies_by_top_10_directors)

plt.title('Number of Movies Directed by Top 10 Directors')
plt.xlabel('Director')
plt.ylabel('Number of Movies')

plt.xticks(rotation=45)

plt.show()
```

```python
highest_duration_movies = df.sort_values(by='duration', ascending=False).head(10)
lowest_duration_movies = df.sort_values(by='duration', ascending=True).head(10)

plt.bar(highest_duration_movies['title'], highest_duration_movies['duration'])
plt.title('Duration of the Highest Duration Movies on Prime Video')
plt.xlabel('Movie Title')
plt.ylabel('Duration')

plt.xticks(rotation=45)

plt.show()
```

```python
plt.bar(lowest_duration_movies['title'], lowest_duration_movies['duration'])
plt.title('Duration of the Lowest Duration Movies on Prime Video')
plt.xlabel('Movie Title')
plt.ylabel('Duration')

plt.xticks(rotation=45)

plt.show()
```
